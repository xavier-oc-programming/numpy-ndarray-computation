## **4)—Manipulating Images as ndarrays**

### **Images as Numerical Data**

An image is nothing more than a collection of pixels.
Each pixel is a combination of **red, green, and blue (RGB)** values, and each of these values is just a number.

This means images can be represented and manipulated as **NumPy ndarrays**.

---

### **Import Statements**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import misc
from PIL import Image
```

* **NumPy** handles numerical operations
* **Matplotlib** displays images
* **SciPy (misc)** provides a sample image
* **PIL** allows loading image files from disk

---

### **Loading and Displaying an Image**

```python
img = misc.face()
plt.imshow(img)
```

Matplotlib interprets the array values as RGB colour values and renders the image accordingly.

---

### **Inspecting the Image Array**

```python
type(img)
```

```python
img.shape
```

```python
img.ndim
```

* Data type: `numpy.ndarray`
* Shape: `(768, 1024, 3)`
* Dimensions: `3`

This means:

* Height: 768 pixels
* Width: 1024 pixels
* Channels: 3 (Red, Green, Blue)

Each colour channel is a 2D matrix, and they are stacked together along the third axis.

---

### **Converting an Image to Grayscale**

To convert an RGB image to grayscale, we use the standard luminance formula:

[
Y = 0.2126R + 0.7152G + 0.0722B
]

This formula assumes RGB values are in the range **0 to 1**, so we must first normalize the image.

---

### **Step 1: Convert to sRGB Range**

```python
sRGB_array = img / 255
```

NumPy uses **broadcasting** to divide every pixel value by 255.

---

### **Step 2: Apply the Grayscale Formula**

```python
grey_vals = np.array([0.2126, 0.7152, 0.0722])
img_gray = sRGB_array @ grey_vals
```

* The `@` operator performs matrix multiplication
* Each pixel’s RGB values are combined into a single grayscale value

---

### **Step 3: Display the Grayscale Image**

```python
plt.imshow(img_gray, cmap='gray')
```

The `cmap='gray'` parameter is essential.
Without it, Matplotlib will apply a default colour map instead of showing a true grayscale image.

---

### **Image Manipulation Challenges**

Because images are ndarrays, we can manipulate them using array operations.

---

### **Flipping an Image Upside Down**

```python
plt.imshow(np.flip(img_gray), cmap='gray')
```

* `np.flip()` reverses the order of rows and columns
* This flips the image vertically and horizontally

---

### **Rotating an Image**

```python
plt.imshow(np.rot90(img))
```

* `np.rot90()` rotates the array 90 degrees counterclockwise
* Works directly on the underlying ndarray

---

### **Inverting (Solarizing) an Image**

To invert colours, subtract pixel values from 255:

```python
solar_img = 255 - img
plt.imshow(solar_img)
```

* Black (0) becomes white (255)
* White becomes black
* NumPy broadcasting applies the operation to every pixel

---

### **Using Your Own Images**

Load an image file and convert it into a NumPy array:

```python
file_name = "yummy_macarons.jpg"
my_img = Image.open(file_name)
img_array = np.array(my_img)
```

Check its structure:

---

### **Key Takeaways**

* Images are just **NumPy ndarrays**
* Each pixel is represented by RGB values
* Broadcasting enables efficient image-wide operations
* Matrix multiplication can be used for colour transformations
* Array manipulation directly translates into visual effects
