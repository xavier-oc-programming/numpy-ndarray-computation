## **5)—Learning Points & Summary**

### **What We Covered in This Lesson**

* Creating NumPy arrays manually using `np.array()`
* Generating arrays programmatically with:

  * `np.arange()`
  * `np.random()`
  * `np.linspace()`
* Inspecting ndarrays by analysing:

  * Shape
  * Dimensions
* Slicing and subsetting ndarrays using indices
* Performing linear-algebra-style operations:

  * Scalar operations
  * Matrix multiplication
* Using NumPy **broadcasting** to make array shapes compatible
* Manipulating images represented as ndarrays

This lesson focused on building a strong mathematical foundation with NumPy.
