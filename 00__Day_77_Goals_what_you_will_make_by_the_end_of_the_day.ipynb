## **0)—Day 77 Goals: what you will make by the end of the day**

### **Overview**

No data science course is complete without **NumPy (Numerical Python)**.

NumPy is a foundational Python library used across **science, engineering, machine learning, data analysis, and computer vision**. It is effectively the **standard for numerical computation in Python**, and many other libraries (such as Pandas, SciPy, and Matplotlib) are built on top of it.

Today’s lesson shifts focus from **high-level data manipulation** (Pandas) to **low-level numerical computation**, where NumPy truly shines.

---

### **Why NumPy Matters**

So far, we’ve been using **Pandas**, which internally relies on NumPy arrays. Pandas is excellent for:

* Time series analysis
* Grouping and aggregation
* Joining and merging datasets
* Handling missing data

However, **NumPy excels at**:

* Fast numerical computation
* Linear algebra
* Multi-dimensional array operations
* Mathematical transformations
* Image and signal processing

Think of it this way:

* **Pandas** → high-level data manipulation
* **NumPy** → low-level numerical power

---

### **Real-World Case Studies**

NumPy is used in some of the most important scientific breakthroughs:

* **First image of a black hole**

  * NumPy (together with SciPy and Matplotlib) enabled processing massive numerical datasets from the Event Horizon Telescope.

* **Detection of gravitational waves**

  * NumPy was used by scientists at LIGO to confirm Einstein’s prediction over 100 years later.

These examples show why learning NumPy is essential.
